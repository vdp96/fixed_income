## Extrapolating interest rates using Nelson Siegel Svensson Model

```python
# fed data
yield_file = os.path.join(DATA_DIR, "feds200628.csv")
fed_df = pd.read_csv(yield_file, skiprows=9)
fed_df["Date"] = pd.to_datetime(fed_df["Date"])
fed_df.head()
```

```python
req_cols = ["Date", "BETA0", "BETA1", "BETA2", "BETA3", "TAU1", "TAU2"]

fed_df = fed_df[req_cols]
fed_df.head()
```

```python
def get_interest_rate_NSS(b0: float, b1: float, b2: float, b3: float, tau1: float, tau2: float, t: float) -> float:
    e = np.exp
    
    t1 = t/tau1
    t2 = t/tau2
    
    rt = b0 + (b1 * (1 - e(-t1))/t1) + (b2 * ((1 - e(-t1))/t1 - e(-t1))) + (b3 * ((1 - e(-t2))/t2 - e(-t2)))
    return rt
```

```python
fed_df["r2"] = fed_df.apply(lambda x: get_interest_rate_NSS(*x[1:7], 2), axis=1)
fed_df["r1"] = fed_df.apply(lambda x: get_interest_rate_NSS(*x[1:7], 2 - 7/365), axis=1)
fed_df["r10"] = fed_df.apply(lambda x: get_interest_rate_NSS(*x[1:7], 10), axis=1)
fed_df["r9"] = fed_df.apply(lambda x: get_interest_rate_NSS(*x[1:7], 10 - 7/365), axis=1)
fed_df["r1w"] = fed_df.apply(lambda x: get_interest_rate_NSS(*x[1:7], 7/365), axis=1)
fed_df.head()
```

```python
plt.figure(figsize=(12, 6))

plt.plot(fed_df["Date"], fed_df["r2"], label="2 Yr Rate")
plt.plot(fed_df["Date"], fed_df["r10"], label="10 Yr Rate")

plt.legend(loc="best")
plt.title("Historical Interest Rates using NSS Model")
plt.xlabel("Year")
plt.ylabel("Interest Rate (%)")

plt.show()
```

```python
ir_df = fed_df[["Date", "r1w", "r2", "r1", "r10", "r9"]]
ir_df = ir_df.set_index("Date")
ir_df.tail(20)
```

## Steeping Based Strategy

```python
START_DATE = "1983-12-30"
END_DATE = "2023-12-31"
INITIAL_CAPITAL = 1000000
MARGIN_REQUIREMENT = 0.1

t_delta = 7/365
```

```python
weekly_dates = pd.date_range(start=START_DATE, end=END_DATE, freq="7D")
df = ir_df[START_DATE:END_DATE]
df = df.resample(rule="7D", label="left").first()
df /= 100
df.head()
```

```python
def cash_position(long_value, short_value, capital):
    total_cash = short_value + capital - long_value
    return total_cash

def get_qty(long_price, short_price, hedge_ratio, capital, margin):
    leverage = 1/margin
    
    position_cost = long_price + hedge_ratio * short_price
    units = capital * leverage / position_cost
    return units

def bond_price(t, ytm, c, fv, n_coupons):
    annuity = fv * ((c/n_coupons)/(ytm/n_coupons)) * (1 - 1/(1 + ytm/n_coupons)**(n_coupons*t))
    
    price = annuity + fv/(1 + ytm/n_coupons)**(n_coupons*t)
    return price

def bond_price_ZCB(t, ytm, fv):
    price = fv/(1 + ytm)**t
    return price

def convexity_ZCB(t, ytm):
    c = t * (t+1)/(1 + ytm)**2
    return c

def duration_ZCB(t, ytm, p):
    return t
```

```python
df["d2"] = 2/(1 + df["r2"])
df["d10"] = 10/(1 + df["r10"])

df["c2"] = df.apply(lambda x: convexity_ZCB(2, x["r2"]), axis=1)
df["c10"] = df.apply(lambda x: convexity_ZCB(10, x["r10"]), axis=1)

df["p2"] = df.apply(lambda x: bond_price_ZCB(2, x["r2"], 100), axis=1)
df["p10"] = df.apply(lambda x: bond_price_ZCB(10, x["r10"], 100), axis=1)

df["DV01_2"] = df["d2"] * df["p2"] * 0.0001
df["DV01_10"] = df["d10"] * df["p10"] * 0.0001
df["hedge_ratio"] = df["DV01_10"]/ df["DV01_2"]
df
```

```python
t2_end = 2 - t_delta
t10_end = 10 - t_delta
df["p2_end"] = (df.shift(-1)).apply(lambda x: bond_price_ZCB(t2_end, x["r1"], 100), axis=1)
df["p10_end"] = (df.shift(-1)).apply(lambda x: bond_price_ZCB(t10_end , x["r9"], 100), axis=1)
df = df.reset_index()
df.head()
```

## Initiate Strategy

```python
df["cash"] = INITIAL_CAPITAL

p2 = df.loc[0, "p2"]
p10 = df.loc[0, "p10"]
hedge_ratio =  df.loc[0, "hedge_ratio"]

initial_cash = df.loc[0, "cash"]

qty = get_qty(long_price=p10, short_price=p2, hedge_ratio=hedge_ratio, capital=initial_cash, margin=MARGIN_REQUIREMENT)
net_cash_position = qty * (-hedge_ratio * p2 + p10) + initial_cash

df.loc[0, "qty"] = qty
df.loc[0, "net_cash_position"] = net_cash_position

position_profit = - (df.loc[0, "p10_end"] - df.loc[0, "p10"]) * qty + (df.loc[0, "p2_end"] - df.loc[0, "p2"]) * qty * hedge_ratio
df.loc[0, "position_profit"] = position_profit

interest = (7/365) * net_cash_position * df.loc[0, "r1w"]
df.loc[0, "interest_accrued"] = interest

net_profit = interest + position_profit
df.loc[0, "net_profit"] = net_profit

final_cash = initial_cash + position_profit + interest
df.loc[1, "cash"] = final_cash

df.head()
```

```python
for i, row in df[1:-1].iterrows():
    p2 = df.loc[i, "p2"]
    p10 = df.loc[i, "p10"]
    hedge_ratio =  df.loc[i, "hedge_ratio"]

    initial_cash = df.loc[i, "cash"]
    
    qty = get_qty(long_price=p10, short_price=p2, hedge_ratio=hedge_ratio, capital=initial_cash, margin=MARGIN_REQUIREMENT)
    net_cash_position = qty * (-hedge_ratio * p2 + p10) + initial_cash

    df.loc[i, "qty"] = qty
    df.loc[i, "net_cash_position"] = net_cash_position

    position_profit = - (df.loc[i, "p10_end"] - df.loc[i, "p10"]) * qty + (df.loc[i, "p2_end"] - df.loc[i, "p2"]) * qty * hedge_ratio
    df.loc[i, "position_profit"] = position_profit

    interest = (7/365) * net_cash_position * df.loc[i, "r1w"]
    df.loc[i, "interest_accrued"] = interest
    
    net_profit = interest + position_profit
    df.loc[i, "net_profit"] = net_profit

    final_cash = initial_cash + position_profit + interest
    df.loc[i+1, "cash"] = final_cash
```

```python
df["weekly_return"] = (df["net_profit"]/df["cash"])
df["cum_return"] = (df["weekly_return"] + 1).cumprod() - 1
df
```

## (i) Cumulative Returns of Steepening Strategy

```python
plt.figure(figsize=(12, 6))

plt.plot(df["Date"], df["cum_return"], label="Cumulative Return")

plt.legend(loc="best")
plt.title("Cumulative Performance of Steepening Strategy")
plt.xlabel("Year")
plt.ylabel("Cumulative Return")

plt.show()
```

## (ii) Convexity Risk of the Spread - (1M USD position in 10 yr bond, 2yr bond is size according to hedge ratio)

```python
cdf = df[["Date", "c2", "c10", "p2", "p10", "hedge_ratio"]]
cdf["qty"] = 1e6/cdf["p10"]
cdf
```

```python
cdf["convexity_risk"] = 0.5 * (0.001)**2 * (hedge_ratio * cdf["qty"] * cdf["p2"] * cdf["c2"] - cdf["c10"] * 1e6)
```

```python
plt.figure(figsize=(12, 6))

plt.plot(cdf["Date"], cdf["convexity_risk"], label="Convexity Risk")

plt.legend(loc="best")
plt.title("Convexity Risk of Steepening Strategy for 1M position in Back Leg")
plt.xlabel("Year")
plt.ylabel("Risk")

plt.show()
```

## (iii) Return Decomposition

```python
df["dr2"] = df["r2"] - df["r1"]
df["dr10"] = df["r10"] - df["r9"]
df
```

```python
df["spread_return"] = (- (df["DV01_2"] * df["hedge_ratio"] * df["dr2"] * 100 - df["DV01_10"] * df["dr10"] * 100) * df["qty"])
```

```python
df["convexity_return"] = (0.5 * df["qty"] * (df["c2"] * df["hedge_ratio"] * df["p2"] * df["dr2"]**2 - df["c10"] * df["dr10"]**2 * df["p10"]))
```

```python
df["time_return"] = df["interest_accrued"]
```

```python
df["residual_return"] = (df["net_profit"] - (df["spread_return"] + df["convexity_return"] + df["time_return"]))
```

```python
ret_df = df[["Date", "net_profit", "spread_return", "convexity_return", "time_return", "residual_return"]].set_index("Date").iloc[:-1]
ret_df
```

```python
cum_ret = (ret_df.cumsum()/1e6)
plt.figure(figsize=(12, 6))

plt.plot(cum_ret.index, cum_ret["spread_return"], label="Spread Return")
plt.plot(cum_ret.index, cum_ret["convexity_return"], label="Convexity Return")
plt.plot(cum_ret.index, cum_ret["time_return"], label="Time Return")
plt.plot(cum_ret.index, cum_ret["residual_return"], label="Residual Return")
plt.plot(cum_ret.index, cum_ret["net_profit"], label="Cumulative Return")
plt.title("Return Decomposition for Steepening Strategy")
plt.ylabel("Return")
plt.legend(loc="best")
plt.show()
```

## Return Contribution

```python
return_contribution = dict()
total = cum_ret.iloc[-1, 0]
spread = cum_ret.iloc[-1, 1]
convexity = cum_ret.iloc[-1, 2]
time = cum_ret.iloc[-1, 3]
residual = cum_ret.iloc[-1, 4]

return_contribution["Spread"] = spread/total * 100
return_contribution["Convexity"] = convexity/total * 100
return_contribution["Time"] = time/total * 100
return_contribution["Residual"] = residual/total * 100
```
